```python
fifa['home_team'] = fifa['home_team'].str.title()
fifa['away_team'] = fifa['away_team'].str.title()
```

```python
fifa['home_team'].unique
```

```python
fifa['home_team'] = fifa['home_team'].str.strip()
fifa['away_team'] = fifa['away_team'].str.strip()
```

```python
fifa[fifa['away_team'] == 'Morocco']
```

```python
# Mapping out Team Form
fifa['home_form'] = fifa['Result'].map({2:3, 0: 0, 1:1})
fifa['away_form'] = fifa['Result'].map({2:0, 0: 3, 1:1})
```

```python
fifa.head()
```

```python
fifa.head()
```

```python
fifa.columns.to_list()
```

```python
training_features = [
 'home_team',
 'away_team',
 'home_score',
 'away_score',
 'Result',
 'home_form',
 'away_form']
# output_feature = ['result']
```

```python
train_fifa = fifa[training_features]
```

```python
train_fifa['home_avg_goal'] = train_fifa.groupby('home_team')['home_score'].transform(
    lambda x: x.shift(1).rolling(window=5).mean()
)
train_fifa['away_avg_goal'] = train_fifa.groupby('away_team')['away_score'].transform(
    lambda x: x.shift(1).rolling(5).mean()
)
train_fifa['home_avg_form'] = train_fifa.groupby('home_team')['home_form'].transform(
    lambda x: x.shift(1).rolling(5).mean()
)
train_fifa['away_avg_form'] = train_fifa.groupby('away_team')['away_form'].transform(
    lambda x: x.shift(1).rolling(5).mean())
```
